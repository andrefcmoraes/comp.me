Sistema Predador Presa EDO

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def predpres(y,t):
    a = 1
    c = 4
    g = 2
    y0=y[0]
    y1=y[1]
    dy0 = y0*(a-a*y1)
    dy1 = y1*(g*y0-c)
    return [dy0,dy1]
```

```python
t = np.linspace(0,10,10001)
yv = [5,1]

y = odeint(predpres,yv,t)
print(y)
```

```python
plt.subplot(1,2,1)
plt.plot(t,y)
plt.subplot(1,2,2)
plt.plot(y[:,0],y[:,1])
plt.show()
```
